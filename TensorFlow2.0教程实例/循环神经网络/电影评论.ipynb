```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
(train_x,train_y),(test_x,test_y)=imdb.load_data()
print(train_x.shape)
print(test_y.shape)
print(len(train_x[1]))
```

```python
word_list=imdb.get_word_index()
print(word_list)
```

```python
word_list={k:(v+3) for k,v in word_list.items()}
word_list["<PAD>"]=0
word_list["<START>"]=1
word_list["<UNK>"]=2
word_list["<EOF>"]=3
word_list.pop("<EOF>")
word_list["<END>"]=3
reverse_word_list={k:v for v,k in word_list.items()}
print(reverse_word_list)
# 13336145618
```

```python
def decode_review(text_list):
    return ' '.join([reverse_word_list.get(i,"<UNK>") for i in text_list])
print(decode_review(train_x[3]))
```

```python
# 对句子中大于某个长度值的进行裁剪，对小于长度值的进行填充
train_x=tf.keras.preprocessing.sequence.pad_sequences(train_x,
                                                      value=0, 
                                                      padding='post',
                                                    truncating='post',
                                                     maxlen=500
                                                     )
```

```python
test_x = np.concatenate(test_x, axis=0)
test_x=tf.keras.preprocessing.sequence.pad_sequences(test_x,
                                                      padding='post',
                                                    truncating='post',
                                                     maxlen=500
                                                     )
print(test_x[1])
```

```python
import numpy as np
train_y=np.array(train_y).reshape(-1,1)
print(train_y.shape)
print(train_x.shape)
```

```python

model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(len(word_list),16,input_length=500))
model.add(tf.keras.layers.GlobalAvgPool1D())
model.add(tf.keras.layers.Dense(64,activation="relu"))
model.add(tf.keras.layers.Dense(2,activation="sigmoid"))
model.summary()
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.sparse_categorical_crossentropy,
             metrics=["accuracy"])
trian_history=model.fit(train_x,train_y,batch_size=32, 
                          epochs=10,
                        validation_split=0.2,
                        validation_freq=20)
print(trian_history)
```

```python
print(model.trainable_variables)
```

```python

model.evaluate(test_x)
```
