```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
w_to_id=[chr(i) for i in np.arange(97,97+26)]
training_set_scaled=np.arange(26)
w_to_id=dict(zip(w_to_id,training_set_scaled))
train_x=[]
train_y=[]
for i in range(4, 26):
    train_x.append(training_set_scaled[i - 4:i])
    train_y.append(training_set_scaled[i])
train_x=np.reshape(train_x,(len(train_x),4))
train_y=np.array(train_y)
print(train_x)
print(len(train_x))
print(train_y)
print(train_x.shape)
```

```python
np.random.seed(7)
np.random.shuffle(train_x)
np.random.seed(7)
np.random.shuffle(train_y)
tf.random.set_seed(7)
```

```python
# model = tf.keras.Sequential([
#     tf.keras.layers.Embedding(26, 2),
#     tf.keras.layers.SimpleRNN(10),
#     tf.keras.layers.Dense(26, activation='softmax')
# ])
model=tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(26,2,))
model.add(tf.keras.layers.SimpleRNN(5))
model.add(tf.keras.layers.Dense(units=26,activation="softmax"))
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)
            ,metrics=["accuracy"]
             )
checkpoint_save_path="./字符预测/Embedding/many_time_step"
if os.path.exists(checkpoint_save_path+".index"):
    print("-----load model-----")
    model.load_weights(checkpoint_save_path)
many_time_step_callbacks=tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_save_path,
                                                  save_best_only=True,
                                                 save_weights_only=True,
                                                 monitor="loss")
train_history=model.fit(train_x,train_y,
                          batch_size=32, 
                          epochs=100, 
                          callbacks=[many_time_step_callbacks])
print(train_history)
```

```python
inputwords = "abcdefghijklmnopqrstuvwxyz"
pre_num=eval(input())
for i in range(pre_num):
    alphet=input("请输入字母:")
    alphet=[w_to_id[ch] for ch in alphet]
    alphet=np.reshape(alphet,(1,4))
    prediction=model.predict(alphet)
#     print(prediction)
    print("预测值："+inputwords[np.argmax(prediction)])
```

```python
with open("./字符预测/Embedding/weights.txt","w") as fp:
    for v in model.trainable_variables:
        fp.write(str(v.name)+"\n")
        fp.write(str(v.shape)+"\n")
        fp.write(str(v.numpy())+"\n")
```
