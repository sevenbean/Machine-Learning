```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
inputword="abcde"
w_to_id={"a":0,"b":1,"c":2,"d":3,"e":4}
train_x=[w_to_id["a"],w_to_id["b"],w_to_id["c"],w_to_id["d"],w_to_id["e"]]
train_y=[w_to_id["b"],w_to_id["c"],w_to_id["d"],w_to_id["e"],w_to_id["a"]]
train_y=np.array(train_y)
train_x=np.reshape(train_x,(len(train_x),1))
print(train_y)
print(train_x)
print(train_y.shape)
print(train_x.shape)
```

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(5,2))
model.add(tf.keras.layers.SimpleRNN(5))
model.add(tf.keras.layers.Dense(units=5,activation="softmax"))
checkpoint_save_path="./字符预测/Embedding_checkpoint_save1"
if os.path.exists(checkpoint_save_path+".index"):
    print("----load model---")
    model.load_weights(checkpoint_save_path)
pre_one_char_callbacks=tf.keras.callbacks.ModelCheckpoint(filepath=
                                                          checkpoint_save_path,
                                                          save_best_only=True,
                                                          save_weights_only=True,
                                                          monitor='loss')
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False)
              ,metrics=["accuracy"]
             )
train_history = model.fit(train_x,train_y,
                          batch_size=32, 
                          epochs=100, 
                          callbacks=[pre_one_char_callbacks])
print(train_history)
```

```python
# 进行预测：
pre_num=eval(input())
for i in range(pre_num):
    alphet=input("请输入字母:")
    alphet=w_to_id[alphet]
    alphet=np.reshape(alphet,(1,1))
    prediction=model.predict(alphet)
    print("预测值："+inputword[np.argmax(prediction)])
```

```python
# 参数的提取
with open("./字符预测/weights_Embedding1.txt","a") as fp:
    for r in model.trainable_variables:
        fp.write(str(r.name)+"\n")
        fp.write(str(r.shape)+"\n")
        fp.write(str(r.numpy())+"\n")
```
