```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Model
from sklearn import datasets
x_data=datasets.load_iris().data
y_data=datasets.load_iris().target
print(x_data.shape)
print(y_data.shape)
```

```python
#自定义网络
class MyModel(Model):
    def __init__(self):
        super(MyModel,self).__init__()
        self.d1=Dense(units=3,activation="softmax",kernel_regularizer=tf.keras.regularizers.l2())
    def call(self,x):
        y=self.d1(x)
        return y
model=MyModel()
model.compile(optimizer=tf.keras.optimizers.Adam(0.1),loss=tf.keras.losses.sparse_categorical_crossentropy,metrics=["accuracy"])
train_history=model.fit(x_data,y_data,verbose=2,epochs=500,validation_split=0.2,validation_freq=20)
print(train_history)
```

```python
import numpy as np
print(y_data[1])
np.random.seed(15)
np.random.shuffle(x_data)
np.random.seed(15)
np.random.shuffle(y_data)
predictions=model.predict(x_data)
print(y_data[1],np.argmax(predictions[1]))
```
