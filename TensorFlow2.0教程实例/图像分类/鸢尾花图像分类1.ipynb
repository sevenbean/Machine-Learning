```python
from sklearn import datasets
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
x_data=datasets.load_iris().data
y_data=datasets.load_iris().target
print(x_data)
print(y_data)
```

```python

model=tf.keras.Sequential()
# model.add(tf.keras.layers.Dense(units=32,input_dim=4,activation="relu"))
model.add(tf.keras.layers.Dense(units=3,activation="softmax",kernel_regularizer=tf.keras.regularizers.l2()))
model.compile(optimizer=tf.keras.optimizers.Adam(0.1),loss=tf.keras.losses.sparse_categorical_crossentropy,metrics=["accuracy"])
train_history=model.fit(x_data,y_data,verbose=2,epochs=500,validation_split=0.2,validation_freq=20)
```

```python
print(train_history.history["loss"])
print(train_history.history["accuracy"])
```

```python

plt.figure(figsize=(10,10))
plt.plot(train_history.history["loss"],label="loss")
plt.plot(train_history.history["accuracy"],label="accuracy")
plt.legend(loc="upper right")
plt.show()
```

```python
prediction=model.predict(x_data)
```
