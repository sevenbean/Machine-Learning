```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16],"ro")
plt.ylabel("same numbers")
plt.show()
```

```python
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei'] # 步骤一(替换sans-serif字体)
plt.rcParams['axes.unicode_minus'] = False   # 步骤二(解决坐标轴负数的负号显示问题)//原文出自【易百教程】，商业转载请联系作者获得授权，非商业请保留原文链接：https://www.yiibai.com/matplotlib/matplotlib_simple_plot.html
x=np.arange(0,np.math.pi*2,0.05)
y=np.sin(x)
plt.plot(x,y,color="green",linestyle='--')
plt.xlabel(u"角度")
plt.ylabel("正弦")
plt.title("正弦波")
plt.show()
```

```python
x=np.linspace(-3,3,20)
y1=2*x+1
y2=x**2
# plt.xlim(-5,5)
plt.figure(figsize=(8,8))
#plt.xticks(np.linspace(-3,3,10))
# plt.yticks([1,3,5,7],["a","b","c","d"])
# 获取当前图片的位置
ax=plt.gca()
#将右边的边框颜色设置为none，默认是白色的
ax.spines["right"].set_color("none")
#将顶部的边框颜色设置为none，默认是白色的
ax.spines["top"].set_color("none")
#将底部的边框往上移到y=3的位置
ax.spines["bottom"].set_position(("data",0))
#将底部的边框往上移到x=0的位置
ax.spines["left"].set_position(("data",0))
plt.plot(x,y1,label="square line",linestyle='--')
plt.plot(x,y2,label="linear line")
plt.legend(loc="upper right")
plt.show()
```

```python
# 散点图
x=np.random.normal(0,1,1024)
y=np.random.normal(0,1,1024)
T=np.arctan2(y,x)
plt.scatter(x,y,c=T)
plt.show()
```

```python
# 柱状图
X=np.arange(12)
y1=(1-X/float(12))*np.random.uniform(0.5,1.0,12)
y2=(1-X/float(12))*np.random.uniform(0.5,1.0,12)
for x, y in zip(X, y1):
    # ha: horizontal alignment
    # va: vertical alignment
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')

for x, y in zip(X, y2):
    # ha: horizontal alignment
    # va: vertical alignment
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va='top')
plt.bar(X,y1)
plt.bar(X,-y2)
plt.xticks([])
plt.yticks([])
ax=plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
plt.show()
```
